```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load the dataset
df = pd.read_csv('Alcohol_Sales.csv', index_col='DATE', parse_dates=['DATE'])
```

```python

# Plot the original time series
plt.plot(df.index, df['S4248SM144NCEN'])
plt.title('Original Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

```python

# Split the data into training and testing sets
train_size = int(len(df) * 0.8)
train_data, test_data = df[0:train_size], df[train_size:len(df)]

# Fit the ARIMA model
model = ARIMA(train_data, order=(5,1,0))
model_fit = model.fit()

# Print the summary of the model
print(model_fit.summary())
```

```python

# Forecast future values
forecast = model_fit.forecast(steps=30)

# Plot the forecasted values
plt.plot(df.index, df['S4248SM144NCEN'])
# Use forecast directly instead of forecast[0] to plot all forecasted values
plt.plot(pd.date_range(start='2019-01-01', periods=30, freq='M'), forecast)
plt.title('Forecasted Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```
